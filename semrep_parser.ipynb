```python
from xml.etree.ElementTree import ElementTree
import pymongo
```

```python
from pymongo import MongoClient
client = MongoClient()
client = MongoClient('localhost', 27017)
```

```python
db = client.semrepData
```

```python
def parse_semrep_xmp(file_name,PMID):
    collection = db['semrep' + str(PMID)]
    semrep_output = ElementTree().parse(file_name)
    for document in semrep_output:
        for semrep_tag in document:
            for utterance in semrep_tag:
                if utterance.tag == 'Predication':
                    predicate = {'type' : 'Predicate'}
                    for pred in utterance:
                        if pred.tag == 'Subject':
                            predicate['Subject'] = pred.attrib
                        if pred.tag == 'Predicate':
                            predicate['Predicate'] = pred.attrib
                        if pred.tag == 'Object':
                            predicate['Object'] = pred.attrib
                        print(pred.tag)
                    collection.insert_one(predicate)
                else:
                    entity = utterance.attrib
                    entity['type'] = 'Entity' 
                    collection.insert_one(utterance.attrib)
                    print(utterance.attrib)
    
```

```python
file_name = 'semrep_output.xml'
parse_semrep_xmp(file_name,223344)
```

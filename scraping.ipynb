```python
import nltk
```

```python
def search(query,max_doc):
    Entrez.email = 'siddakid007@gmail.com'
    handle = Entrez.esearch(db='pubmed', 
                            sort='relevance', 
                            retmax=str(max_doc),
                            retmode='xml', 
                            term=query)
    results = Entrez.read(handle)
    return results
```

```python
#verb=GetRecord&identifier=oai:pubmedcentral.nih.gov:4304705&metadataPrefix=pmc
def fetch_details(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'siddakid007@gmail.com'
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids,
                          verb='GetRecord',
                          identifier='oai',
                          metadataPrefix='pmc')
    results = Entrez.read(handle)
    return results
```

```python
def fetch_abstract(idx):
    Entrez.email = 'siddakid007@gmail.com'
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=str(idx))
    results = Entrez.read(handle)
    try:
        abstract=results['PubmedArticle'][0]['MedlineCitation']['Article']['Abstract']['AbstractText']
    except:
        pass
        return ''
    else:
        ab_string=''
        for ab in abstract:
            ab_string+=(ab+'\n')
        return ab_string
```

```python
def ie_preprocess(document):
    sentences = nltk.sent_tokenize(document)
    sentences = [nltk.word_tokenize(sent) for sent in sentences]
    sentences = [nltk.pos_tag(sent) for sent in sentences]
    return sentences
```

```python
def doc_tokenize(document):
    sentences = nltk.sent_tokenize(document)
    sentences = [nltk.word_tokenize(sent) for sent in sentences]
    return sentences
```
